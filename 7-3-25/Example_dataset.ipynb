```python
import pandas as pd
from sklearn.linear_model import LinearRegression
df=pd.read_csv('headbrain.csv')
```

```python
df.head()
```

```python
x=df['Head Size(cm^3)']
type(x)
```

```python
x=df[['Head Size(cm^3)']]
type(x)
```

```python
y=df['Brain Weight(grams)']
type(y)
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
m=model.coef_[0]
c=model.intercept_
print(f'slope of model:{m:.2f}')
print(f'intercept of model:{c:.2f}')
```

```python

from sklearn.metrics import r2_score
y_predict=model.predict(x)
r2_score=r2_score(y,y_predict)
print(f'{r2_score:.2f}')
```
