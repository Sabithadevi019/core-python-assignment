```python
import pandas as pd
data=pd.read_csv('job.csv')
data
```

```python

data.columns
```

```python
df=pd.DataFrame(data)
df
```

```python
df.dtypes
```

```python
# Check for NaN (missing values) in all columns.
df.isnull().sum()
```

```python
# Fill missing values using mode (for categorical) and median (for numeric) where appropriate. 
for column in df.columns:
    if df[column].dtype == 'object': 
        mode_value = df[column].mode()[0] 
        df[column].fillna(mode_value, inplace=True)
    elif df[column].dtype in ['int64', 'float64']: 
        median_value = df[column].median()
        df[column].fillna(median_value, inplace=True)
print(df.isna().sum()) 
```

```python
df
```

```python
# Convert date, StartDate, OpeningDate, Deadline into datetime format
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['StartDate'] = pd.to_datetime(df['StartDate'], errors='coerce')
df['OpeningDate'] = pd.to_datetime(df['OpeningDate'], errors='coerce')
df['Deadline'] = pd.to_datetime(df['Deadline'], errors='coerce')
df.dtypes
```

```python
#  Days left to apply → Deadline - OpeningDate. 
df['StartDate'] = pd.to_datetime(df['StartDate'], errors='coerce')
df['OpeningDate'] = pd.to_datetime(df['OpeningDate'], errors='coerce')
df['Deadline'] = pd.to_datetime(df['Deadline'], errors='coerce')
df['DaysLeftToApply'] = (df['Deadline'] - df['OpeningDate']).dt.days.abs()
df['JobStartDelay'] = (df['StartDate'] - df['OpeningDate']).dt.days.abs()
df[['DaysLeftToApply', 'JobStartDelay']] = df[['DaysLeftToApply', 'JobStartDelay']].fillna(0)
print(df[['DaysLeftToApply','JobStartDelay']])
```

```python
# Extract year & month from dates (Year, Month). 
df['OpeningYear'] = df['OpeningDate'].dt.year
df['OpeningMonth'] = df['OpeningDate'].dt.month
df['StartYear'] = df['StartDate'].dt.year
df['StartMonth'] = df['StartDate'].dt.month
df['DeadlineYear'] = df['Deadline'].dt.year
df['DeadlineMonth'] = df['Deadline'].dt.month
print(df[['OpeningDate', 'OpeningYear', 'OpeningMonth', 'StartDate', 'StartYear', 'StartMonth', 'Deadline', 'DeadlineYear', 'DeadlineMonth']])
```

```python
# Split Location into City & Country.
df['CommaCount'] = df['Location'].str.count(',')
df[['City', 'Country']] = df['Location'].str.split(',', n=1, expand=True)
df['City'] = df['City'].str.strip()
df['Country'] = df['Country'].str.strip()
print(df[['Location','City','Country']])
```

```python
# Identify IT-related jobs from the IT column.
it_jobs = df[df['IT'] == True]
it_jobs
```

```python
# Categorize Eligibility and Audience into structured Education Level / Experience Ranges.
def categorize_eligibility(row):
    if "qualified" in row:
        return "No specific qualifications required"
def categorize_audience(row):
    if "interested" in row:
        return "Open to all"
df['Eligibility_Categorized'] = df['Eligibility'].apply(categorize_eligibility)
df['Audience_Categorized'] = df['Audience'].apply(categorize_audience)
df[['Eligibility_Categorized','Audience_Categorized']]
```

```python
# Removing HTML tags, special characters, and excessive whitespace for job requirement,job description
import re
def clean_text(text):
    text = re.sub(r'<.*?>', '', text) 
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text) 
    text = re.sub(r'\s+', ' ', text).strip() 
    return text
df['Cleaned_JobDescription'] = df['JobDescription'].apply(clean_text)
df['Cleaned_JobRequirment'] = df['JobRequirment'].apply(clean_text)
print(df[['Cleaned_JobDescription', 'Cleaned_JobRequirment']].head())
```

```python
#  Tokenizing and lowercasing for NLP tasks. 
import re
def tokenize(text):
    tokens = text.lower().split()
    return ' '.join(tokens)
df['Cleaned_JobDescription'] = df['JobDescription'].apply(tokenize)
df['Cleaned_JobRequirment'] = df['JobRequirment'].apply(tokenize)
print(df[['Cleaned_JobDescription', 'Cleaned_JobRequirment']].head())
```
