**Data visualization**

```python

plt.figure(figsize=(8, 5))
plt.hist(data1["Age"], bins=10, color="skyblue", edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

**Boxplot**

```python
plt.figure(figsize=(6, 5))
sns.boxplot(y=data1["Salary"], color="lightcoral")
plt.title("Salary Distribution with Outliers")
plt.ylabel("Salary")
plt.show()
```

**Scatter plot**

```python
plt.figure(figsize=(8, 5))
plt.scatter(data1["Experience"], data1["Salary"], alpha=0.6, color="purple")
plt.title("Experience vs Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

**Bar plot**

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="Department", y="Salary", hue="Department", data=data1, palette="coolwarm", legend=False)
plt.title("Average Salary per Department")
plt.xlabel("Department")
plt.ylabel("Salary")
plt.xticks(rotation=45)
plt.show()
```

**Pair plot**

```python
sns.pairplot(data1[["Age", "Salary", "Experience"]], diag_kind="kde")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(data1.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```
