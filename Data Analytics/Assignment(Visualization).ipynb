```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df=pd.read_excel('Desktop\\Sales.xlsx')
```

```python
df
```

```python
plt.figure(figsize=(17,5))
plt.plot(df['Month'],df['Sales'],color='green',linestyle='dashed',marker='*',linewidth=2)
plt.title('Growth of sales')
plt.xlabel('Months(1-12)')
plt.ylabel('Sales')
plt.show()
```

```python
import pandas as pd
df=pd.read_excel('Desktop\\Order.xlsx')
df
```

```python
plt.hist(df['Quantity'],bins=5,color='blue',edgecolor='black')
plt.title('Order Quantity Distribution')
plt.xlabel('Order ID')
plt.ylabel('Quantity')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
df=pd.read_excel('Desktop\\Age.xlsx')
df

```

```python
plt.hist(df['Age'],bins=6,color='yellow',edgecolor='black')
plt.title('Visualization of Age')
plt.xlabel('Age')
plt.ylabel('Number of Persons')
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['Age'],df['Spending Amount'],color='pink',s=100,edgecolor='black')
plt.title('Scatter Plot Example')
plt.xlabel('Age')
plt.ylabel('Spening Amount')
plt.show()
```

```python
import pandas as pd
df=pd.read_excel('Desktop\\Student.xlsx')
df

```

```python
y=df[['Math','Science','English','History']].mean()
y.plot(kind='bar',color=['orange','green','pink','yellow'])
plt.title('Basic Bar Chart')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```

```python
y=df[['Math','Science','English','History']].boxplot()
plt.title('Basic Bar Chart')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```
