```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('train.csv')
df
```

```python
sns.barplot(data=df,x='Pclass',y='Survived',palette={1:'lightblue',2:'pink',3:'yellow'})
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
sns.scatterplot(data=df,x='Age',y='Fare',hue='Survived',style='Embarked',palette='magma')
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
y=df[['Age','Fare','Pclass','Survived']]
sns.pairplot(y,hue='Survived',palette='magma')
plt.suptitle('Age vs Fare')

plt.show()
```

```python
df_numeric = df.select_dtypes(include='number')
corre = df_numeric.corr()
sns.heatmap(corre, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
