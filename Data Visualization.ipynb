```python
X=[1,2,3,4,5]
Y=[2,4,6,8,10]
plt.plot(X,Y);
plt.title("Simple Line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot(X,Y,color='pink',linestyle='dashed',marker='*',linewidth=2)
plt.title('Customized Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
x=[1,2,3,4,5,6]
y=[1000,1500,2000,2500,3000,3500]
plt.plot(x,y)
plt.plot(x,y,color='green',linestyle='dashed',marker='*',linewidth=2)
plt.title('Growth of savings')
plt.xlabel('Months(1-6)')
plt.ylabel('Savings amount')
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color='black')
plt.title('Basic Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
plt.bar(categories,values,color=['green','grey','pink','blue','red'],edgecolor=['darkgreen','black','pink','darkblue','orange'],width=0.5)
```

```python
ages=[20,32,38,43,56,58,60,64,69,70]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color='pink',edgecolor='black',width=5)
plt.title('Age Distribution of Customers')
plt.xlabel('Age ranges')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
x=['Python','Java','C++','Ruby','DBMS']
y=[20,30,15,35,43]
plt.pie(y,labels=x,colors=['yellow','pink','green','blue','red'],autopct='%1.1f%%',startangle=90)
plt.title('Percentage of Languages')
plt.legend()
```

```python
x=[5,10,15,20,25]
y=[10,20,30,40,50]
plt.scatter(x,y,color='green',s=100,edgecolor='black')
plt.title('Scatter Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```
